```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("../data/processed/team_games_features.csv")

# Quick sanity check
print(df.columns)
print(df.head())
```

```python
sns.barplot(data=df, x="b2b", y="point_diff")
plt.title("Average Point Differential: Back-to-Back vs Non-B2B")
plt.xlabel("Back-to-Back (1 = Yes, 0 = No)")
plt.ylabel("Average Point Differential")
plt.show()
```

```python
df.groupby("b2b")["point_diff"].agg(["count", "mean"])
```

```python
df["b2b"].value_counts()
```

```python
df["rest_bucket"] = pd.cut(
    df["rest_days"],
    bins=[-1,1,2,10],
    labels=["0-1 Days","2 Days","3+ Days"]
)

df.groupby("rest_bucket")["point_diff"].agg(["count","mean"])
```

```python
sns.barplot(data=df, x="rest_bucket", y="point_diff")
plt.title("Average Point Differential by Rest Days")
plt.xlabel("Rest Days")
plt.ylabel("Average Point Differential")
plt.show()
```
